> **Примітка:** Грузія - це країна з *надзвичайною культурною спадщиною* і *дружніми людьми*. Вона славиться своєю *винами* та *гостинністю*.


```python
def gnome_sort_desc(lst):
    i = 0
    while i < len(lst):
        if i == 0 or lst[i] <= lst[i - 1]:
            i += 1
        else:
            lst[i], lst[i - 1] = lst[i - 1], lst[i]
            i -= 1
    return lst

def task_8():
    
    input_list = [5, 3, 78, 3, 23, 7, 34]
    
    sorted_list = gnome_sort_desc(input_list)
    print(sorted_list)

task_8()
```
